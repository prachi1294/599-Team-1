```python
#use spacy sentence to find entities
display_key = True

class entities:
    def __init__(self, name, count,linenumber,label):
        self.name = name
        self.count = count
        self.lineappear = [0,linenumber]
        self.label = label
    def addonce(self,linenumber):
        self.count = self.count + 1
        self.lineappear.append(linenumber)

entities_list=[]
entities_count=[]
entities_objects=[]
test_sentence=sentence[100:110]
#test_sentence=sentence

linenumber=0
for sent in sentence:
    linenumber=linenumber+1
    print('')
    print(linenumber)
    print(sent)
    doc=sent
    if display_key:
        #print(doc)
        print(doc.ents)
        print("")
    for ent in doc.ents:
        temp_ent=str(ent)
#         print(temp_ent)
#         print((temp_ent in entities_list))
        if (temp_ent in entities_list):
            ent_index=entities_list.index(temp_ent)
            entities_count[ent_index]=entities_count[ent_index]+1
            entities_objects[ent_index].addonce(linenumber)
        else:
            entities_list.append(temp_ent)
            entities_count.append(1)
            entities_objects.append(entities(temp_ent,1,linenumber,ent.label_))
```
